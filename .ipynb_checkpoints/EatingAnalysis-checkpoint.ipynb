```python

def getRestData(df):
    resturants = []
    pre_res = ""
    for index, row in df.iterrows():
        if row[5] != pre_res:
            if resturants and not resturants[-1][1]: resturants.pop()
            resturants.append((row[5],[]))
        if row[3] > 0: continue
        pre_res = row[5]
        resturants[-1][1].append(row[2])

    return resturants
    
```

```python
import numpy as np

def getAve(data, timeInterval):
    total = np.sum(data)
    return int(total/timeInterval)

def plot(data, title = "Heart Rate"):
    plt.plot(data)
    plt.ylabel(title)
    plt.show()
```

```python
## Get all the values without time interval

totalData = []
for index, row in df.iterrows():
    if row[3] > 0: continue
    totalData.append(row[2])

plot(totalData, "heart rate")


## Get the average value for every a specific number of minutes
start = 0
aveCol = []
subsection = []
interval = 10

for index, row in df.iterrows():
    if start == interval:
        aveCol.append(getAve(subsection, interval))
        start = 0
        subsection.clear()
    else:
        start += 1
        subsection.append(row[2])

plot(aveCol, "heart rate")
```

```python
def getVar(data):
    return np.var(data)




def getArrayAverages(data, interval = 10):
    start = 0
    aveCol = []
    subsection = []
    for val in data:
        if start == interval:
            aveCol.append(getAve(subsection, interval))
            start = 0
            subsection.clear()
        else:
            start += 1
            subsection.append(val)
    return aveCol
```
