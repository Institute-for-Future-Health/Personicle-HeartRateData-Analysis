```python
model = DataAnalyzer()
model.readfiles("Tools/jordan2_segment.csv")
```

```python
eating_out = model.getEatingSeg()
pre_nonmove_out = model.getPreNonMovingSegmentId()
```

```python
import matplotlib.pyplot as plt


colors = ['blue','orange']
eatings = []
nonmoves = []
for eat,nonmove in zip(eating_out, pre_nonmove_out):
    eatings.append(eat[1])
    nonmoves.append(nonmove[1])
    
plt.plot(eatings,c = colors[0], label=colors[0])
plt.plot(nonmoves,c = colors[1], label=colors[1])
plt.ylabel("Comparison of average")
print("Blue: Eating Segments", "Orange: previous nonmovement segments")
plt.show()
```

```python
df = pd.read_csv("Datasets/jordan2_label.csv", encoding="ISO-8859-1")

data = []
for index, row in df.iterrows():
    if data and row[4] == data[-1]['food']:
        data[-1]['heartrate'].append(row[2])
    else:
        data.append({'segid' : row[0], 'heartrate' : [row[2]], 'food' : row[4]})

print(len(data))
```

```python
for item in data:
    print(item['food'], len(item['heartrate']))
    plt.plot(item['heartrate'])
    plt.show()
    print('\n')
```
