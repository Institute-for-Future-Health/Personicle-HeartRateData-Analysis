```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')

data = pd.read_csv('eating_jordan1.csv')
print(data.shape)
data.head()
```

```python
def getRestData(df):
    resturants = []
    pre_res = ""
    for index, row in df.iterrows():
        if row[5] != pre_res:
            if resturants and not resturants[-1][1]: resturants.pop()
            resturants.append((row[5],[]))
        if row[3] > 0: continue
        pre_res = row[5]
        if row[2]>0:
            resturants[-1][1].append(row[2])
    return resturants

def plot(data, title = "Heart Rate"):
    plt.plot(data)
    plt.ylabel(title)
    plt.show()
```

```python
resturants = getRestData(data)
for item in resturants:
    print(item[0])
    plot(item[1])
    print("\n")
```
