```python
from Tools.SugarTest import *
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp

from scipy import signal
%load_ext autoreload
%autoreload 2
```

```python
ms_data = getMsData()
```

```python
print(ms_data)
```

```python
rawdata = getTimeInfo()
print(rawdata)
```

```python
fitbit_data = getFitbitData(rawdata)
```

```python
total_array = fitbit_data + ms_data
```

```python

label = ["low sugar", "high sugar"]
for item in total_array:
    display_data = sp.signal.medfilt(item['heartrates'])
    plt.plot(display_data)
    plt.show()
    print(label[item['sugarlevel']])
    print("Foodname" ,item['foodname'])
    print('\n')
```

```python
high_sugar = []
low_sugar = []

for item in total_array:
    if item['sugarlevel']:
        high_sugar.append(item['heartrates'])
    else:
        low_sugar.append(item['heartrates'])

print(len(high_sugar), len(low_sugar))
```

```python
high_sugar_ave = np.mean(high_sugar, axis= 0)
low_sugar_ave = np.mean(low_sugar, axis= 0)
print(len(high_sugar_ave), len(low_sugar_ave))
```

```python
print("high sugar")
plt.plot(high_sugar_ave)
plt.show()
print('\n')

print("low sugar")
plt.plot(low_sugar_ave)
plt.show()
```
